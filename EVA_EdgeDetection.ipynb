```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25

#kernel to show vertical edges
kernel = np.float32([[-2,4,-2],[-2,4,-2],[-2,4,-2]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
kernelone = np.float32([[-2,-2,-2],[4,4,4],[-2,-2,-2]])

#kernel to show horizontal edges
dstone = cv2.filter2D(image,-1,kernelone)

cv2_imshow(dstone)
```

```
kerneltwo = np.float32([[-4,-4,8],[-4,8,-4],[8,-4,-4]])

#kernel to show 45 degree angle edges
dsttwo = cv2.filter2D(image,-1,kerneltwo)
cv2_imshow(dsttwo)
```

```
kernelblur = np.float32([[0.11,0.11,0.11],[0.11,0.11,0.11],[0.11,0.11,0.11]])

#kernel to show blurriness in the edges
dstblur = cv2.filter2D(image,-1,kernelblur)
cv2_imshow(dstblur)
```

```
kernelsharp = np.float32([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

#kernel to show sharpness in the edgdes 

dstsharp = cv2.filter2D(image,-1,kernelsharp)
cv2_imshow(dstsharp)
```

```
kernelid = np.float32([[0,0,0],[0,1,0],[0,0,0]])

#identity function kernel

dstid = cv2.filter2D(image,-1,kernelid)
cv2_imshow(dstid)
```
